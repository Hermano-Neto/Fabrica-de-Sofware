### a) qual a causa mais frequente e a mais rara de acidentes registrados nos dois anos?

passo a passo:

1- Utilizei o mesmo metodo que aprendi na letra C da 2ª tarefa(value_counts()).

2- Criei uma lista para cada ano para inserir o mais frequente e o(s) mais raro(s) caso(s) de acidentes. Os coloquei utilizando o append, o qual apenas coloca o valor ou termo que estiver dentro do ().

3- No ano de 2020 tiveram mais de uma causa rara, entçao tive que criar um metodo que colocassem todos dentro da lista. Este método consiste em uma repetição, a qual verifica se a quantidade de acidentes x é igual ao de y, se for igual ele pegara o valor y e ira inserir na lista, se não, irá parar a repetição.

4- Criei um dataframe que contem as causas mais raras e frequentes de 2019 e 2020, utilizando o mesmo método da 2ª tarefa letra C.

```python
causa1 = df1['causa_acidente'].value_counts()
l1_4a = [] 
        
l1_4a.append(causa1.index[0])
l1_4a.append(causa1.index[-1])
l1_4a
```

```python
causa2 = df2['causa_acidente'].value_counts()
l2_4a = []

l2_4a.append(causa2.index[0])
l2_4a.append(causa2.index[-1])

c = -1
while True:
    if causa2[c] == causa2[c - 1]:
        l2_4a.append(causa2.index[c - 1])   
        c -= 1
        
    else:
        break
l2_4a
```

```python
df_4a = pd.DataFrame({'Causa do acidente': [l1_4a[1], l2_4a[1], l2_4a[2], l2_4a[3], l2_4a[4], l2_4a[5], l2_4a[6], l2_4a[7], l2_4a[8],
                                l2_4a[9], l2_4a[10]]},
                     index = [['2019', '2020', '2019', '2020', '2020', '2020', '2020', '2020', '2020', '2020', '2020'],
                              ['mais frequente', 'mais frequente', 'mais rara', 'mais rara', 'mais rara', 'mais rara',
                               'mais rara', 'mais rara', 'mais rara', 'mais rara', 'mais rara']])

df_4a.index.names = ['Ano', 'Classificação']
df_4a
```

### c) mostre, reorganizando o conjunto de dados, se a pandemia conseguiu diminuir ou não a incidência de acidentes nas rodovias federais.

passo a passo:

1- Em relação a quantidade de acidentes, calculei a diferença em quantidade e porcentagem entre os anos. O método que utilizei foi verificar a quantia de linhas utilizando o len() e depois fiz as contas necessárias.

2- Em relação a quantidade de acidentes por estado, calculei a diferença em quantidade e porcentagem entre os anos. O método utilizado foi a partir do value_counts(), que conta a repetição de palavras, depois coloquei em uma lista apenas os estados em que houve redução de acidentes. Para isso tive de criar duas repetições, as quais verificaraim se o conteudo de uf 2019 é igual a uf de 2020, ja que as series de cada ano não ficaram na mesma ordem.

3- Utilizei o mesmo esquema do passo 1 para descobrir a diferença e porcentagem de pessoas e mortos.

4- Criei um dataframe em q as colunas eram as porcentagens e a diferença, as linhas as classificações de acidentes que utilizei e os estados oi país e depois coloquei nomes para as linhas.

```python
acidente1 = len(df1.index)
acidente2 = len(df2.index)

diferença_acidentes = acidente1 - acidente2

porcentagem_acidentes = f'{1 - acidente2 / acidente1:.1%}'

print(diferença_acidentes)
print(porcentagem_acidentes)
```

```python
estados1 = df1['uf'].value_counts()
estados2 = df2['uf'].value_counts()

l1_4c = []
l2_4c = []
for i in range(27):
    for j in range(27):
        if  estados1.index[i] == estados2.index[j] and estados1[i] - estados2[j] > 0:
            l1_4c.append(estados1.index[i])
            l1_4c.append(estados1[i] - estados2[j])

        if estados1.index[i] == estados2.index[j] and estados2[i] / estados1[i] < 1:
            l2_4c.append(estados1.index[i])
            l2_4c.append(f'{(1 - estados2[i] / estados1[i]):.1%}')
            
print(l1_4c)
print()
print(l2_4c)
```

```python
pessoas1 = df1.pessoas.sum()
pessoas2 = df2.pessoas.sum()

diferença_pessoa = pessoas1 - pessoas2
porcentagem_pessoa = f'{(1 - pessoas2 / pessoas1):.1%}'

print(diferença_pessoa)
print(porcentagem_pessoa)
```

```python
mortos1 = df1.mortos.sum()
mortos2 = df2.mortos.sum()

diferença_mortos = mortos1 - mortos2
porcentagem_mortos = f'{(1 - mortos2 / mortos1):.1%}'

print(diferença_mortos)
print(porcentagem_mortos)
```

```python
df_4c = pd.DataFrame({'Porcentagem': [porcentagem_acidentes, l2_4c[1], l2_4c[3], l2_4c[5], l2_4c[7], l2_4c[9], l2_4c[11],
                                      l2_4c[13], l2_4c[15], l2_4c[17], l2_4c[19], l2_4c[21], l2_4c[23], l2_4c[25], l2_4c[27], 
                                      l2_4c[29], l2_4c[31], l2_4c[33], porcentagem_pessoa, porcentagem_mortos],
                      
                      'Diferença': [diferença_acidentes, l1_4c[1], l1_4c[3], l1_4c[5], l1_4c[7], l1_4c[9], l1_4c[11] ,
                                      l1_4c[13], l1_4c[15], l1_4c[17], l1_4c[19], l1_4c[21], l1_4c[23], l1_4c[25], 
                                      l1_4c[27], l1_4c[29], l1_4c[31], l1_4c[33], diferença_pessoa, diferença_mortos]},
                     
                     index = [['Geral', 'Estado', 'Estado', 'Estado', 'Estado', 'Estado', 'Estado', 'Estado', 'Estado',
                               'Estado', 'Estado', 'Estado', 'Estado', 'Estado', 'Estado', 'Estado', 'Estado', 'Estado',
                               'Pessoas', 'Mortos'],
                    
                              
                              ['Brasil', l1_4c[0], l1_4c[2], l1_4c[4], l1_4c[6], l1_4c[8], l1_4c[10], l1_4c[12], l1_4c[14],
                               l1_4c[16], l1_4c[18], l1_4c[20], l1_4c[22], l1_4c[24], l1_4c[26], l1_4c[28], l1_4c[30],
                               l1_4c[32], 'Brasil', 'Brasil']])

df_4c.index.names = ['Acidentes', 'Local']
df_4c
```
