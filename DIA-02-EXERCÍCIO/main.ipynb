```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv("netflix1.csv")
```

```python
df.head()
```

```python
plt.hist(x=df['type'], color= 'red');
```

A analise levou em consideração apenas 2 itens, sendo eles movie e tv show, os quais representam os tipos de vídeos que existem na netflix.
  O filme(movie) é o que mais aparece, com cerca de 66% do conteudo analisado da netflix, enquanto as series de tv(tv shows) possui apenas 33%. 

```python
df_release_year = pd.DataFrame(df.release_year.value_counts(normalize = False)).sort_index(ascending = True)
df_release_year
```

```python
plt.plot(df_release_year);
```

  O grafico indica a quantidade de filmes, os quais estão na netflix, que foram feitos no ano.
  Fazendo a analise, nota-se que a partir do anos 2000(100), existe um grande crescimento na quantidade de filmes produzidos. Possuindo um crescimento ainda maior nos anos de 2012(200) ate 2018(1150). Porém, no ano de 2019 existe uma leve queda e no de 2020(650) uma enorme queda, tendo uma produção com quase a metade do ano de 2018.
